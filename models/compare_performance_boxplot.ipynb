```python
# do not delete! this is for plotting accuracy, recall and fscore. recall had an error, thats why we didnt include it.
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSVs
cnn_data = pd.read_csv('cnn_test_performance_per_fold_4HOLDS.csv')
rf_data = pd.read_csv('rf_test_performance_per_fold_4HOLDS.csv')

# Add a column to identify the model
cnn_data['Model'] = 'CNN'
rf_data['Model'] = 'RF'

# Combine the two dataframes
data = pd.concat([cnn_data, rf_data], ignore_index=True)

# Prepare the data for plotting
melted_data = pd.melt(data, id_vars=['K-Fold', 'Model'], value_vars=[
    'CNN_Accuracy', 'CNN_Recall', 'CNN_FScore', 
    'RF_Accuracy', 'RF_Recall', 'RF_FScore'],
    var_name='Metric', value_name='Value')

# Map the Metric names to a simpler form
metric_mapping = {
    'CNN_Accuracy': 'Accuracy', 'CNN_Recall': 'Recall', 'CNN_FScore': 'F-Score',
    'RF_Accuracy': 'Accuracy', 'RF_Recall': 'Recall', 'RF_FScore': 'F-Score'
}
melted_data['Metric'] = melted_data['Metric'].map(metric_mapping)

# Setup the aesthetic style of the plots
sns.set(style="whitegrid", palette="pastel")

# Create a color palette that distinguishes between CNN and RF
palette = {"CNN": "skyblue", "RF": "lightgreen"}

# Create the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=melted_data, x='Metric', y='Value', hue='Model',
            palette=palette, linewidth=1.5, fliersize=3)

# Enhancements for better visualization
plt.title('Model Performance Metrics Comparison')
plt.xlabel('M  E  T  R  I  C', labelpad=20)
plt.ylabel('P E R F O R M A N C E')
plt.legend(title='Model', loc='upper right')

# Show the plot
plt.show()
```
